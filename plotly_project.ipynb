```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
```

```python
data = pd.read_csv("marketing_and_product_performance.csv")
```

```python
data.head()
```

```python
data.info()
```

```python
fig1 = px.scatter(
    data,
    x="Budget",
    y="ROI",
    size="Units_Sold",
    color="Subscription_Tier",
    hover_data=["Campaign_ID", "Revenue_Generated", "Product_ID"],
    title="ROI vs Budget (Bubble Size = Units Sold)"
)
fig1.show()
```

**total revenue by supscription**

```python
revenue_by_tier = data.groupby("Subscription_Tier")["Revenue_Generated"].sum().reset_index()

fig2 = px.bar(
    revenue_by_tier,
    x="Subscription_Tier",
    y="Revenue_Generated",
    color="Subscription_Tier",
    title="Total Revenue by Subscription Tier"
)
fig2.show()
```

**customer satsfaction level vs discout level**

```python
fig3 = px.box(
    data,
    x="Discount_Level",
    y="Customer_Satisfaction_Post_Refund",
    color="Subscription_Tier",
    title="Customer Satisfaction vs Discount Level"
)
fig3.show()
```

```python
fig4 = px.scatter(
    data,
    x="Clicks",
    y="Conversions",
    color="Subscription_Tier",
    size="Revenue_Generated",
    hover_data=["Campaign_ID", "Product_ID"],
    title="Conversions vs Clicks"
)
fig4.show()
```

```python
keywords_counts = data["Common_Keywords"].value_counts().reset_index()
keywords_counts.columns = ["Keyword", "Count"]

fig7 = px.pie(
    keywords_counts,
    values="Count",
    names="Keyword",
    title="Distribution of Common Keywords"
)
fig7.show()
```
